Google Trends Link: https://trends.google.com/trends/
Pytrends Link: https://pypi.org/project/pytrends/

## Variables - 

### 1. Average resale price (StockX) – Dependent Variable
### 2. Retail price (StockX or the original product listing) – Control Variable
### 3. Trend Demand (Google Trends + StockX)
### Calculation – Trend Demand = Google search interest/Number of resale listings
The Trend Demand variable measures how big consumer interest is compared to the supply in the resale market. The higher the value the higher the demand, based on supply. 

## Research Question -
**Question: Do fashion trends influence resale prices in secondary markets for fashion products?** 

##Background & Motivation -
Resale fashion is a large part of the fashion economy now. Companies like StockX or GOAT allow consumers to buy/trade sneakers, streetwear items, and luxury items. These resale markets tend to have higher prices for items than the original retail price. 
Where consumers attention is can influence demand and pricing of items. Google trends allow us to see consumers attention through online search behavior. Increased search interest could lead to rising demand for an item. 
This project will combine the market data and search interest data to analyze whether consumer interest influences resale prices. Knowing more about this relationship can help understand digital consumer behavior, prices in marketplaces, and demand signals.
